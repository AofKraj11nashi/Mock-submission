## Entity relationship diagram

User authentication (diagram created with miro.com)



#  1. Log In Screen (Entity)

## Attributes
| Attribute | Type | Description |
|----------|------|-------------|
| screen_id (PK) | Integer | Unique identifier for this screen entity |
| email (NOT NULL) | Varchar(100) | User's registered email |
| password (NOT NULL) | Varchar(100) | Users password |

## Relationships
- Connects to **Sign Up Screen**.
- Connects to **Forgot Password Screen**.
- On success  **Profile Page**.

---

#  2. Sign Up Screen (Entity)

## Attributes
| Attribute | Type | Description |
|----------|------|-------------|
| screen_id (PK) | Integer | Unique identifier |
| full_name (NOT NULL) | Varchar(100) | Users full name |
| email (NOT NULL) | Varchar(100) | New account email |
| password (NOT NULL) | Varchar(100) | Account password |

## Relationships
- Connects to **Log In Screen**.
- Successful sign-up  **Profile Page**.
- Connects to **Forgot Password Screen**.

---

#  3. Forgot Password Screen (Entity)

## Attributes
| Attribute | Type | Description |
|----------|------|-------------|
| screen_id (PK) | Integer | Unique identifier |
| email (NOT NULL) | Varchar(100) | Email used for password recovery |

## Relationships
- Returns to **Log In Screen** after reset.
- Connects to **Sign Up Screen**.

---

#  4. Profile Page (Entity)

## Attributes
| Attribute | Type | Description |
|----------|------|-------------|
| screen_id (PK) | Integer | Unique identifier |
| profile_picture | Varchar(200) | File path or image URL |
| full_name (NOT NULL) | Varchar(100) | Users display name |
| email (NOT NULL) | Varchar(100) | Registered email |
| pronouns | Varchar(50) | Optional field |
| phone_number | Varchar(15) | Optional field |
| skills_hobbies | Varchar(255) | Users declared skills |

## Relationships
- Receives successful login from **Log In Screen**.
- Receives new account from **Sign Up Screen**.
